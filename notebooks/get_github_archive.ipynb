```python
import requests
import boto3
```

```python
filename = '2025-10-03-10.json.gz'
r = requests.get(f'https://data.gharchive.org/{filename}', stream=True)
```

```python
r.raise_for_status()
```

```python
s3_client = boto3.client(
    's3',
    endpoint_url='http://localhost:9000',
    aws_access_key_id='admin',
    aws_secret_access_key='password123'
)
bucket_name = 'github-events'
object_key = f"raw/2025/10/03/{filename}"
s3_client.upload_fileobj(r.raw, bucket_name, object_key)
```
